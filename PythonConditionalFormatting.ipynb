```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'Value': [10, 15, 13, 17, 20, 23, 21, 25, 27, 30]
}

# Create a DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Define a function to apply conditional formatting
def formatDataPointColor(val):
    if val < 15:
        return 'green'
    elif val < 25:
        return 'orange'
    else:
        return 'red'
```

```python
# Apply the conditional formatting to the data points
df['Color'] = df['Value'].apply(formatDataPointColor)
```

```python
df.head()
```

```python
# Create a seaborn line chart
def ShowConditionalFormatting():
  line_chart = sns.lineplot(
    x='Year',
    y='Value',
    data=df,

    # Set line color
    color='blue',

    # Set line width
    linewidth=2,

    # Set line style
    linestyle='--',

    # Set marker style, size and transparency
    marker='o',
    markersize=8,
    alpha=0.8
)

  # Add conditional formatting to the data points
  for i in range(len(df)):
      plt.plot(
          df['Year'][i],
          df['Value'][i],
          marker='o',
          markersize=10,
          markerfacecolor=df['Color'][i],
          markeredgewidth=2,
          markeredgecolor='black'
          )

  # Set chart title and labels
  plt.title('Sample Line Chart with Conditional Formatting', fontsize=12)
  plt.xlabel('Year', fontsize=10)
  plt.ylabel('Value', fontsize=10)

  # Show the plot
  plt.show()
```

```python
ShowConditionalFormatting()
```
