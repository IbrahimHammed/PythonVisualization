```python
# dummy dataset
years = [2010, 2011, 2012, 2013, 2014, 2015]
west = [10, 14, 17, 18, 21, 27]
east = [13, 15, 19, 23, 26, 30]
```

```python
# this produces a blank canvas
def CreateBlankCanvas():
    fig, ax = plt.subplots()
    return fig, ax
```

```python
CreateBlankCanvas()
```

```python
# Adding a flexible chart title to the blank canvas
def AddTitle(title):
    fig, ax = CreateBlankCanvas()
    ax.set_title(title, fontsize="small", color="navy")
    return fig, ax
```

```python
AddTitle("Sales Analysis")
```

```python
def AddAxisLabels():
    fig, ax = AddTitle("Sales Analysis")
    ax.set_title("Sales Analysis", fontsize="small", color="navy")
    ax.set_xlabel("SalesYear", fontsize="x-small", color="#A70D83")
    ax.set_ylabel("Sales (millions)", fontsize="x-small", color="#A70D83")
    return fig, ax
```

```python
AddAxisLabels()
```

```python
# Set limits on both axis
def AddAxisLimit():
    fig, ax = AddAxisLabels()
    ax.set_xlim(2010, 2015)
    ax.set_ylim(0, 40)
    return fig, ax
```

```python
AddAxisLimit()
```

```python
# Increase figure size
def SetChartSize():
    fig, ax = AddAxisLimit()
    fig.set_size_inches(7, 5)
    return fig, ax
```

```python
SetChartSize()
```

```python
# adds data to the updated canvas
def AddData(x, y1, y2):
    fig, ax = SetChartSize()
    #ax.plot(x, y1, ls=":", c="green")
    #ax.plot(x, y2, ls="--", c="red")

    ax.plot(x, y1, ls="dotted", c="green")
    ax.plot(x, y2, ls="dashed", c="red")

    return fig, ax
```

```python
AddData(years, east, west)
```

```python
# Add legends to distinguish between the data series
def AddLegends(x, y1, y2):
    fig, ax = AddData(x, y1, y2)
    # Option-1
    ax.legend(["East", "West"], loc="lower right")

    # Possible location:
    #"center", "best"
    # "upper right", "upper left", "upper center"
    #"lower right", "lower left", "lower center"
    #"right", "center left", "center right"

    return fig, ax
```

```python
AddLegends(years, east, west)
```

```python
# Add legends to distinguish between the data series
def AddLegendsMethod2(x, y1, y2):
    fig, ax = AddData(x, y1, y2)
    # Option-2
    ax.legend(["East", "West"], bbox_to_anchor=(1.2, 1))

    return fig, ax
```

```python
AddLegendsMethod2(years, east, west)
```

```python
# Markers
def AddMarkers(x, y1, y2):
    fig, ax = AddLegends(x, y1, y2)
    ax.plot(x, y1, ls="dotted", c="green", marker="o", markersize=5)
    ax.plot(x, y2, ls="dashed", c="red", marker="*", markersize=10)
    return fig, ax
```

```python
AddMarkers(years, east, west)
```

```python
def AddTicks(x, y1, y2):
    fig, ax = AddMarkers(x, y1, y2)
    plt.xticks(rotation=45)
    return fig, ax
```

```python
AddTicks(years, east, west)
```

```python
# Vertical lines
```

```python
def AddVerticalLine(x, y1, y2):
    fig, ax = AddTicks(x, y1, y2)
    ax.axvline(2012, ls="--", c="black")
    return fig, ax
```

```python
AddVerticalLine(years, east, west)
```

```python
# Horizontal lines
```

```python
def AddHorizontalLine(x, y1, y2):
    fig, ax = AddVerticalLine(x, y1, y2)
    ax.axhline(26, ls="--", c="red")
    return fig, ax
```

```python
AddHorizontalLine(years, east, west)
```

```python
# Text
```

```python
def AddText(x, y1, y2):
    fig, ax = AddHorizontalLine(x, y1, y2)
    ax.text(2014, 20, "Your key message!", fontsize="x-small", color="navy")
    return fig, ax
```

```python
AddText(years, east, west)
```

```python
# Annotations
def AddAnnotations(x, y1, y2):
    fig, ax = AddText(x, y1, y2)
    ax.annotate("Your message", xy=(2012,15), xytext=(2012,15), arrowprops={"facecolor":"red"})

    return fig, ax
```

```python
AddAnnotations(years, east, west)
```

```python
# Combined Chart Features
def CombinedChartFeatures(chartTitle,xLabel, yLabel, xLowerLimit, xUpperLimit, yLowerLimit, yUpperLimit, TransactionYears, WestSales, EastSales, ChartLength, ChartHeight):
    fig, ax = plt.subplots()

    # Add chart title
    ax.set_title(chartTitle, fontsize="small", color="red")

    # Add Axis Labels
    ax.set_xlabel(xLabel, fontsize="x-small", color="black")
    ax.set_ylabel(yLabel, fontsize="x-small", color="black")

    # Adjust chart size
    fig.set_size_inches(ChartLength, ChartHeight)

    # Set limits on both axis
    ax.set_xlim(xLowerLimit, xUpperLimit)
    ax.set_ylim(yLowerLimit, yUpperLimit)

    # Add Data
    ax.plot(TransactionYears, WestSales, ls="dotted", c="green", marker="o", markersize=5)
    ax.plot(TransactionYears, EastSales, ls="dashed", c="red", marker="*", markersize=10)

    # Add Legends
    ax.legend(["East", "West"], bbox_to_anchor=(1.2, 1))

    # Add Ticks
    plt.xticks(rotation=45)

    # Add vertical lines
    ax.axvline(2014, ls="--", c="grey", alpha= 0.3)


```

```python
CombinedChartFeatures("Annual Sales Tracker", "TransactionYear", "Sales ($ millions)", 2010, 2015, 0, 40, years, west, east, 8, 7)
```

```python
# connect and load external dataset
url = "https://github.com/IbrahimHammed/PythonVisualizations/blob/main/SalesData.xlsx?raw=true"
dfSales = pd.read_excel(url)
```

```python
# Preview it
dfSales.head()
```

```python
def LineChart():
    fig, ax = plt.subplots()

    # Syntax is like ax.plot(x-axis data, y-axis data, decorations)
    ax.plot(
        dfSales.SalesYear,
        dfSales.Total,
        marker="*"
        )

    ax.plot(
        dfSales.SalesYear,
        dfSales.East,
        marker="*"
        )
    return fig, ax
```

```python
LineChart()
```

```python
# Stacked Line Chart (Area)
```

```python
def StackedLineChart():
    fig, ax = plt.subplots()

    #
    ax.stackplot(
        dfSales.SalesYear,
        dfSales.West,
        dfSales.East,
        dfSales.Total
        )

    return fig, ax
```

```python
StackedLineChart()
```

```python
# Dual Axis Chart
```

```python
def DoubleAxisChart():
    fig, ax = plt.subplots()

    #
    ax.plot(
        dfSales.SalesYear,
        dfSales.West,
        label="West",
        marker="*",
        markersize=10,
        linestyle="dashed"
        )

    # Pay attention to this line
    ax2 = ax.twinx()

    ax2.plot(
        dfSales.SalesYear,
        dfSales.Total,
        label="Total",
        marker="o",
        markersize=10,
        linestyle=":",
        color="orange"
        )

    return fig, ax
```

```python
DoubleAxisChart()
```

```python
def Donut():
    fig, ax = plt.subplots()

    mydataset = [450, 200, 780]

    ax.pie(
        mydataset,
        startangle = 90,
        autopct="%.0f%%"
        )

    hole = plt.Circle((0,0), 0.70, color="white")
    plt.gcf()

    fig.gca().add_artist(hole)

    return fig, ax
```

```python
Donut()
```

```python
def SpecialDonut():
    fig, ax = plt.subplots()

    mydataset = [450, 200, 780]

    ax.pie(
        mydataset,
        startangle = 90,
        pctdistance=0.85,
        colors= ["white", "white", "green"]
        )

    hole = plt.Circle((0,0), 0.70, fc="white")
    plt.gcf()

    fig.gca().add_artist(hole)

    plt.text(0,0,"55%", ha="center", va="center", fontsize=50)

    return fig, ax
```

```python
SpecialDonut()
```
