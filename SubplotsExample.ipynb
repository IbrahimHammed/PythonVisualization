```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Get data
url = "https://github.com/IbrahimHammed/PythonVisualizations/blob/main/ResultMatrix.xlsx?raw=true"
dfExam = pd.read_excel(url)
```

```python
dfExam.head()
```

```python
dfExam['TotalScore'] = dfExam.Maths + dfExam.English + dfExam.Biology + dfExam.Chemistry + dfExam.Physics
```

```python
dfExam.head()
```

```python
# Subplots
def ChartSubplots():
  fig, ax  = plt.subplots(2, 2, constrained_layout=True, figsize = (16, 10))

  fig.suptitle("Subplot Demo", fontsize=14)

  # chart 1: Maths
  ax[0][0].plot(
      dfExam.CandidateID,
      dfExam.Maths,
      label="Maths",
      color="black",
      ls="dashed",
      marker="*",
      markersize=10,
      markeredgecolor="black",
      markerfacecolor="red"
  )

  # Decorate the subplot
  ax[0][0].set_ylim(0, 100)
  ax[0][0].set_xlim(1, 24)
  ax[0][0].set_xlabel("Candidate ID", fontsize=12, color="navy")
  ax[0][0].set_ylabel("Marks (%)", fontsize=12, color="navy")
  ax[0][0].set_title("Maths Marks (%)")


  # chart 2: English
  ax[0][1].plot(
      dfExam.CandidateID,
      dfExam.English,
      label="English",
      color="black",
      ls="dashed",
      marker="*",
      markersize=10,
      markeredgecolor="black",
      markerfacecolor="green"
  )

  # Decorate the subplot
  ax[0][1].set_ylim(0, 100)
  ax[0][1].set_xlim(1, 24)
  ax[0][1].set_xlabel("Candidate ID", fontsize=12, color="navy")
  ax[0][1].set_ylabel("Marks (%)", fontsize=12, color="navy")
  ax[0][1].set_title("English Marks (%)")


  # chart 3: Biology
  ax[1][0].plot(
      dfExam.CandidateID,
      dfExam.Biology,
      label="Biology",
      color="black",
      ls="dashed",
      marker="*",
      markersize=10,
      markeredgecolor="black",
      markerfacecolor="yellow"
  )

  # Decorate the subplot
  ax[1][0].set_ylim(0, 100)
  ax[1][0].set_xlim(1, 24)
  ax[1][0].set_xlabel("Candidate ID", fontsize=12, color="navy")
  ax[1][0].set_ylabel("Marks (%)", fontsize=12, color="navy")
  ax[1][0].set_title("Biology Marks (%)")


  # chart 4: Chemistry
  ax[1][1].plot(
      dfExam.CandidateID,
      dfExam.Chemistry,
      label="Maths",
      color="black",
      marker="*",
      markersize=10,
      markeredgecolor="black",
      markerfacecolor="orange"
  )

  # Decorate the subplot
  ax[1][1].set_ylim(0, 100)
  ax[1][1].set_xlim(1, 24)
  ax[1][1].set_xlabel("Candidate ID", fontsize=12, color="navy")
  ax[1][1].set_ylabel("Marks (%)", fontsize=12, color="navy")
  ax[1][1].set_title("Chemistry Marks (%)")

  return fig, ax
#
```

```python
ChartSubplots()
```

```python
dataX = {
    "Subjects": ["Maths", "English", "Biology", "Chemistry", "Physics"],
    "Total": [MathsTotal, EnglishTotal, BiologyTotal, ChemistryTotal, PhysicsTotal]
}

dataX
```

```python
dfX = pd.DataFrame(dataX)
dfX
```

```python
from types import LambdaType
def ResultPieChart():
    fig, ax = plt.subplots()

    mydataset = dfX.Total

    ax.pie(
        # All the rows must be numeric
        mydataset,

        # Sets start angle for the pie chart
        startangle = 90,

        # formats the display of % share in each segment
        autopct="%.0f%%",

        # defines colours for the different segments
        colors=["green", "pink", "orange", "blue", "yellow"],

        labels = ["Maths", "English", "Biology", "Chemistry", "Physics"],

        # defines which segment blows out based on the number of segments
        explode=[0, 0, 0.1, 0, 0],

        # creates a shadow effect
        shadow=False
        )


    return fig, ax
```

```python
ResultPieChart()
```
